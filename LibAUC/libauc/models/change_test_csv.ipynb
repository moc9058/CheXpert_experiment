```python
import os
import pandas as pd
files_lst = ['CheXpert_origin','CheXpert_male','CheXpert_female','CheXpert_before40','CheXpert_after40']
for file in files_lst:
    test_csv = os.path.join('C:\\','Users','moc90','Library',file,'test.csv')
    before_test_pd = pd.read_csv(test_csv)
    after_test_pd = pd.DataFrame(columns=before_test_pd.columns)
    paths_lst = before_test_pd['Path'].tolist()
    for i in range(len(paths_lst)):
        paths_elements_lst = paths_lst[i].split('/')
        new_path = f"{paths_elements_lst[0]}/test/{paths_elements_lst[2]}/{paths_elements_lst[3]}/{paths_elements_lst[4]}"

        row = before_test_pd.loc[i]
        row_keys = list(row.keys())
        row_values = list(row.values)
        row_values[0] = new_path
        row_pd = pd.DataFrame(data=[row_values],columns=row_keys)
        # print(row_pd)
        after_test_pd = pd.concat([after_test_pd,row_pd],ignore_index=True)
    # print(before_test_pd)
    print(after_test_pd)
    after_test_pd.to_csv(test_csv,index=False)
```
