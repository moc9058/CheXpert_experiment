```python
import os
import pandas as pd

origin_train_valid_test_pd = pd.read_csv(os.path.join(os.getcwd(),'CheXpert_origin','train_valid_test.csv'))

folders_lst = ['CheXpert_origin','CheXpert_male','CheXpert_female','CheXpert_before40','CheXpert_after40']
for folder in folders_lst:
    before_valid_pd = pd.read_csv(os.path.join(os.getcwd(),folder,'valid.csv'))
    before_test_pd = pd.read_csv(os.path.join(os.getcwd(),folder,'test.csv'))
    before_valid_pd.to_csv(os.path.join(os.getcwd(),folder,'valid_backup.csv'),index=False)
    before_test_pd.to_csv(os.path.join(os.getcwd(),folder,'test_backup.csv'),index=False)
    
    after_valid_pd = pd.DataFrame(columns=origin_train_valid_test_pd.columns)
    after_test_pd = pd.DataFrame(columns=origin_train_valid_test_pd.columns)
    for i in range(len(before_valid_pd)):
        row = before_valid_pd.loc[i]
        row_keys = list(row.keys())
        row_values = list(row.values)
        row_values[0] = row_values[0].replace('train','valid')
        row_pd = pd.DataFrame(data=[row_values],columns=row_keys)
        after_valid_pd = pd.concat([after_valid_pd,row_pd],ignore_index=True)
    for i in range(len(before_test_pd)):
        row = before_test_pd.loc[i]
        row_keys = list(row.keys())
        row_values = list(row.values)
        row_values[0] = row_values[0].replace('train','test')
        row_pd = pd.DataFrame(data=[row_values],columns=row_keys)
        after_test_pd = pd.concat([after_test_pd,row_pd],ignore_index=True)
    print(after_valid_pd)
    print(after_test_pd)
    print()
    after_valid_pd.to_csv(os.path.join(os.getcwd(),folder,'valid.csv'),index=False)
    after_test_pd.to_csv(os.path.join(os.getcwd(),folder,'test.csv'),index=False)
```
